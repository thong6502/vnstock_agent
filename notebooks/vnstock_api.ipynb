```python
quote = Quote(symbol='VCI', source='VCI')
```

```python
df = quote.history(start='2025-11-01', end='2025-11-13', interval="1D")
df[(df["time"] >= "2025-11-01") & (df["time"] <= "2025-11-13")]
```

```python
delta = df["close"].diff()
gain = delta.clip(lower=0)
loss = -delta.clip(upper=0)

avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()

rs = avg_gain / avg_loss
rsi = 100 - (100 / (1 + rs))
```

```python
rsi
```

```python
df["SMA2"]=df["close"].rolling(window=2).mean()
```

```python
df[["close", "SMA2"]]
```
